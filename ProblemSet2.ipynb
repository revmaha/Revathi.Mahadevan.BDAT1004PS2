 global a

Initially, the variable a has a value of 0. 
In b(), the variable a is set as a global variable. Here, a is changed to store the value of the function c(). 
Function c() returns the value  of a + 2 which is stored in a in the funtion b().

First call of b():
- a is set as a global variable. (note: a=0)
- a is changed to return value of c(). 
- c() returns a + 2 i.e 0 + 2 = 2.
- The new value of a is 2 after the first call of b()

Second call of b():
- a is set as a global variable. (note: a=2)
- a is changed to return value of c(). 
- c() returns a + 2 i.e 4.
- The new value of a is 4 after the second call of b()

Third call of b():
- a is set as a global variable. (note: a=4)
- a is changed to return value of c(). 
- c() returns a + 2 i.e 6.
- The new value of a is 6 after the third call of b()


Therefore, when print(a) is executed the value of a which is 6 is displayed.

```python
from urllib.parse import urljoin
from urllib.request import urlopen
from html.parser import HTMLParser
class webdir(HTMLParser):

    def __init__(self, url):
        HTMLParser.__init__(self)
        self.url = url
        self.links = []

    
    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    absolute = urljoin(self.url, attr[1])
                        self.links.append(absolute)
                        
    def getLinks(self):
        return self.links

url = "https://reed.cs.depaul.edu/lperkovic/test1.html"
resource = urlopen(url)
content = resource.read().decode()
collector = webdir(url)
collector.feed(content)
collector.getLinks()
```

```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']

# Upper case
words_upper = [words.upper() for words in words]
print(words_upper)

#Lower case
words_lower = [words.lower() for words in words]
print(words_lower)

#Lengths of words in list words
length = []
for each in words:
    leng = len(each)
    length.append(leng)
    
print(length)

#The list containing a list for every word of list words, where each list contains the word in uppercase and lowercase and the length of the word.
new = list(zip(words_upper,words_lower,length))
print(new)

#The list of words in list words containing 4 or more characters
four = []

for each in words:
    if len(each) >= 4:
        four.append(each)
    
print(four)
```
